Przelicz wskazanie czujnika na temperaturę dzieląc przez wartość z tabeli

Wypisz temperaturę z dokładnością do trzeciego miejsca po przecinku (tak jak na przykładzie) i ze znakiem specjalnym ℃ (\u2103)
    
BONUS. Do punktu 1 dodaj walidację - upewnij się, że użytkownik podał jeden z dni który możemy znaleźć w danych

BONUS. Do punktu 4 dodaj wizualizację temperatury w zakresie 0 do 100 stopni na pasku złożonym z 20 znaków (np. dla 50 stopni, zapełnione jest 10/20 znaków)


Wskazanie czujnika - przelicznik:

Większe niż, lub równe 1400 - 22.5

Pomiędzy 1400, a 1200 - 23.1

Mniejsze niż, lub równe 1200 - 23.5

```python
data = """
monday;1250
tuesday;1405
wednesday;1750
thursday;1100
friday;0800
saturday;1225
sunday;1355
"""
```

```python
#funkcja zapisująca daty do slownika z podanego zbioru
def data_to_dict(data):

    data_lines = data.strip().split('\n')
    data_dict = {}

    for line in data_lines:
        dat = line.split(';')
        data_dict[dat[0]] = int(dat[1])

    return data_dict
```

```python
#funkcja pobierajace wartosc dla czytnika
def get_sensor_value(day):
    data_dict = data_to_dict(data)
    
    return data_dict[day]
```

```python
#funkcja przeliczajaca temperature zgodnie z podanymi danymi
def calculate_temp(value):
    if value >= 1400:
        return value/22.5
    elif 1200 < value < 1400:
        return value/23.1
    else:
        return value/23.5
```

```python
#funkcja do narysowania temperatury graficznie
def draw_graph(temperature):
    number_of_hash = int(temperature/100 * 20)
    graph = ''
    for index in range(22):
        if index == 0 or index == 21:
            graph += '|'
        elif index <= number_of_hash:
            graph += '#'
        else:
            graph += '-'
    return graph
```

```python
#generowanie raportu
def generate_raport():
    input_day = input("Proszę o wprowadzenie poprawnego dnia tygodnia: ")
    day_to_check = str(input_day).lower()
    try:
        sensor_value = get_sensor_value(day_to_check)
        temperature_value  = calculate_temp(sensor_value)
        message = f'Dane dla "{day_to_check}"\nWskazanie czujnika: {sensor_value}\nTemperatura: {temperature_value:.03f}\u2103\n0\u2103 '
        message += draw_graph(temperature_value) # wizualizacja grafu - bonus nr 2
        message += ' 100\u2103'
        print (message)
    except:
        print("Proszę o wprowadzenie poprawnego dnia tygodnia: ") # sprawdzenie czy dzień jest poprawny

generate_raport()
```

### Zadanie 3.

Napisz funkcję create_tower, która jako input przyjmuje integera większego od 0
i zwraca listę, która tworzy "piramidę" z tyloma piętrami:
Przykładowe rozwiązanie:
In: create_tower(3)
Out:
[
  '  *  ',
  ' *** ',
  '*****'
]

```python
floors = input("Podaj liczbę całkowitą pięter, większą od 0")

def create_tower(floors):
    try: 
        if int(floors) > 0:
            fl = int(floors)
            for i in range(1, fl + 1):
                print((fl-i)*" "+ i*"*" + (i-1)*"*" + (fl-i)*" ")
                i+=1
        else:
            print("Proszę podać liczbę całkowitą większą od zera") 
    except:
        print("Proszę podać liczbę całkowitą większą od zera")

create_tower(floors)
```

Wyświetlaj aktualny stan planszy przy pomocy znaków 'o', 'x', '-' w trzech wierszach Np.:

--x

-x-

oo-

Naprzemiennie pytaj gdzie postawić 'x', a gdzie 'o' przyjmując numer pozycji od 1 do 9 Tzn.:

123

789

Zakończ grę gdy jedna ze stron wygra, lub wszystkie miejsca będą zapełnione

Na razie wyświetlaj wszystko po kolei, jedno pod drugim

```python
#nadanie tablicy
def initialize_board():
    return {1: ["-","-","-"],
            2: ["-","-","-"],
            3: ["-","-","-"]}

#funkcja wizualizujaca tablice
def visualize_board(board):
    for pos in board:
        row = ''
        for value in board[pos]:
            row += value
        print(row)
    print("")

#funkcja ktora opisuje pozycje - z pozycji okresla kroty rzad i kolumna w naszej tablicy odpowiada lokalizacji
def get_row_column(position):
    row = (position - 1) // 3 + 1
    if row == 1:
        column = position - row
    elif row == 2:
        column = position - 4
    else: 
        column = position - 7
    return [row, column]

#funkcja ktora nadpisuje nam "-" w naszej tablicy na x i o
def insert_symbol_to_position(board, symbol, position):
    indexes = get_row_column(position)
    symbol_to_update = board[indexes[0]][indexes[1]]
    if symbol_to_update == '-':
        board[indexes[0]][indexes[1]] = symbol
        return True
    return False

#funckja definiująca kolejnosc symboli x i o
def get_symbol(round):
    symbol = ''
    if round % 2 == 0:
        symbol += 'x'
    elif round % 2 != 0:
        symbol += 'o'
    return symbol

#funckja weryfikujaca wygrana
def verify_if_won(position, board):
    indexes = get_row_column(position)
    row = indexes[0]
    column = indexes[1]
    if((board[row][0] == board[row][1]) & (board[row][0] == board[row][2])):
        return True
    if((board[1][column] == board[2][column]) & (board[1][column] == board[3][column])):
        return True
    if board[2][1] == '-':
        return False
    if((board[1][0] == board[2][1]) & (board[1][0] == board[3][2])):
        return True
    if((board[1][2] == board[2][1]) & (board[1][2] == board[3][0])):
        return True

def tic_tac_toe():
    board = initialize_board()

    round = 0
    while round < 9:
        try: 
            symbol = get_symbol(round)

            input_position = int(input(f'Where do you want to put "{symbol}": '))

            if input_position < 1 or input_position > 9:
                print('Wrong position')
                continue

            is_inserted = insert_symbol_to_position(board, symbol, int(input_position))
                
            if is_inserted == False:
                print('This place is already taken, try again')
                continue

            won = verify_if_won(int(input_position),board)
            if won:
                visualize_board(board=board)
                print(f'{symbol} won the round')
                break

            round += 1
        except:
            print("Wrong input")

        visualize_board(board=board)

tic_tac_toe()
```
